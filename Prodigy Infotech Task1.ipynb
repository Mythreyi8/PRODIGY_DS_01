```python
import numpy as np
import matplotlib.pyplot as plt 
 
  
# Define age categories
age_categories= ['0-14M', '15-64M', '65M','0-14F', '15-64F', '65F']

# Create a new DataFrame with columns for each age category
df_stacked = df[['Population ages 0-14, male ',' Population ages 15-64, male ','Population ages 65 and above, male',
                 'Population ages 0-14, female ','Population ages 15-64, female ',
                 'Population ages 65 and above, female ']].copy()
df_stacked.columns = age_categories



# Calculate the sum of each age category
total_population = df_stacked.sum()

# Create a bar chart for age distribution
plt.figure(figsize=(30,10))
plt.bar(age_categories,total_population)
plt.xlabel('Age Category')
plt.ylabel('Population Total')
plt.title('Bar Chart for Age and Gender Distribution in the Population (2022)')
plt.show()
```
